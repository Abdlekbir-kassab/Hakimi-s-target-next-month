```python
# استيراد المكتبات - Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1 Define Hakimi’s current stats

# Current player stats
matches = 38  # Number of matches
goals = 6  # Number of goals
assists = 14  # Number of assists

# Calculate offensive contribution per match
contribution_per_match = (goals + assists) / matches

# Display the results
print(f"Matches: {matches}")
print(f"Goals: {goals}")
print(f"Assists: {assists}")
print(f"Contribution per match: {contribution_per_match:.2f}")
```

```python
#2: Set the target improvement for next month

# Define the realistic improvement percentages
goal_increase_percentage = 0.05  # 5% increase in goals (realistic)
assist_increase_percentage = 0.05  # 5% increase in assists (realistic)

# Calculate the expected goals and assists for next month
expected_goals_next_month = goals * (1 + goal_increase_percentage)
expected_assists_next_month = assists * (1 + assist_increase_percentage)

# Display the expected contributions for next month
print(f"Expected goals next month: {expected_goals_next_month:.2f}")
print(f"Expected assists next month: {expected_assists_next_month:.2f}")
```

```python
# 3: Set realistic contribution goal for next month

# Current player stats
current_contributions = 20  # Current total contributions (goals + assists)
matches_played = 38  # Number of matches played so far

# Calculate current contribution rate per match
current_contribution_rate = current_contributions / matches_played

# Number of matches next month
matches_next_month = 4  # Expected matches in next month

# Calculate expected contributions based on current rate
expected_contributions_next_month = current_contribution_rate * matches_next_month

# Set target contributions for next month
target_contributions_next_month = 6  # Goal is to have 6 contributions in the next month

# Calculate the required improvement
required_improvement = target_contributions_next_month - expected_contributions_next_month

# Display results
print(f"Current contributions per match: {current_contribution_rate:.2f}")
print(f"Expected contributions next month: {expected_contributions_next_month:.2f}")
print(f"Target contributions next month: {target_contributions_next_month}")
print(f"Required improvement: {required_improvement:.2f} contributions")
```

```python
# 4: Analyze areas for improvement

# Current player stats
current_goals_per_match = goals / matches  # Average goals per match
current_assists_per_match = assists / matches  # Average assists per match

# Expected improvement (based on target contribution of 6 for 4 matches)
target_goals_per_match = 6 / 4  # Expected goals per match to reach 6 contributions in 4 matches
target_assists_per_match = 6 / 4  # Expected assists per match to reach 6 contributions in 4 matches

# Calculate the improvement required in goals and assists
required_goals_improvement = target_goals_per_match - current_goals_per_match
required_assists_improvement = target_assists_per_match - current_assists_per_match

# Display analysis
print(f"Current goals per match: {current_goals_per_match:.2f}")
print(f"Current assists per match: {current_assists_per_match:.2f}")
print(f"Target goals per match to achieve 6 contributions: {target_goals_per_match:.2f}")
print(f"Target assists per match to achieve 6 contributions: {target_assists_per_match:.2f}")

print(f"Required improvement in goals per match: {required_goals_improvement:.2f}")
print(f"Required improvement in assists per match: {required_assists_improvement:.2f}")
```

```python
import matplotlib.pyplot as plt

# Current player stats
current_goals_per_match = 6 / 38  # Average goals per match
current_assists_per_match = 14 / 38  # Average assists per match

# Target goals and assists per match to achieve 6 contributions in 4 matches
target_goals_per_match = 6 / 4
target_assists_per_match = 6 / 4

# Required improvement
required_goals_improvement = target_goals_per_match - current_goals_per_match
required_assists_improvement = target_assists_per_match - current_assists_per_match

# Plotting the current vs target performance
labels = ['Goals per Match', 'Assists per Match']
current_values = [current_goals_per_match, current_assists_per_match]
target_values = [target_goals_per_match, target_assists_per_match]
improvement_needed = [required_goals_improvement, required_assists_improvement]

x = range(len(labels))

fig, ax = plt.subplots(figsize=(8, 5))

# Plotting the bars for current, target and improvement required
bar_width = 0.3
ax.bar(x, current_values, bar_width, label='Current Performance', color='skyblue')
ax.bar([p + bar_width for p in x], target_values, bar_width, label='Target Performance', color='orange')
ax.bar([p + 2*bar_width for p in x], improvement_needed, bar_width, label='Required Improvement', color='green')

# Adding labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Performance')
ax.set_title('Goals and Assists Performance vs Target')
ax.set_xticks([p + bar_width for p in x])
ax.set_xticklabels(labels)
ax.legend()

# Display the plot
plt.show()
```
